```python
_old_xp = 0.0
_old_yp = 0.0
with open('xpos_ypos_30s.dat','w') as f:
    for xp, yp in zip(xpos,ypos):
        if xp > _old_xp:
            f.write("%f %f\n" % (xp, yp))
        _old_xp = xp
        _old_yp = yp
```
